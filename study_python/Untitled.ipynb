```python
import numpy as np

t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t))
plt.grid()
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t))
plt.grid()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('sin')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t), lw=3, label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.title('sin')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

```python
import pandas as pd

col_names = ['', '', '', '', 
             '', '','', '', '']

raw_data = pd.read_excel("./data/teenage_mental.xls", 
                         header=1, 
                         usecols="C:K", 
                         names=col_names)

raw_data.loc[1] = 100. - raw_data.loc[0]
raw_data[''] = ['','']
raw_data.set_index('',drop=True, inplace=True)

raw_data
```

```python
raw_data[''].plot.pie()
```

```python
raw_data[''].plot.pie(explode = [0, 0.02]);
```

```python
f, ax = plt.subplots(1, 3, figsize=(16, 8))

raw_data[''].plot.pie(explode=[0, 0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('   ')
ax[0].set_ylabel('')

raw_data[''].plot.pie(explode=[0, 0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('   ')
ax[1].set_ylabel('')

raw_data[''].plot.pie(explode=[0, 0.02], ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('   ')
ax[2].set_ylabel('')

plt.show()
```
